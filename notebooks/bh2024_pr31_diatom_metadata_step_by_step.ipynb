```python
from omnipy import (HttpUrlDataset, HttpUrlModel, runtime, PandasDataset, StrDataset, TaskTemplate, JsonDataset)

runtime.config.data.terminal_size_lines = 500
runtime.config.data.terminal_size_columns = 100
```

```python
@TaskTemplate
def create_google_spreadsheet_url_database(title: str, spreadsheet_id: str, google_api_key: str) -> HttpUrlDataset:
    url = HttpUrlModel('https://sheets.googleapis.com/')
    url.path // 'v4' // 'spreadsheets' // spreadsheet_id
    url.query['key'] = google_api_key
    return HttpUrlDataset({title: url})
```

```python
spreadsheet_title = 'Meta-Data 49 Diatom Protein Coding Gene Annotations'
spreadsheet_id = '1_1wKibSXbi8MQCTqpK03ytuQH71yn28xKuRS1i-mVSw'
google_api_key = 'PUT YOU API KEY HERE'
# To get a Google Sheets API key, follow the instructions e.g. here: 
# https://handsondataviz.org/google-sheets-api-key.html
urls = create_google_spreadsheet_url_database.run(spreadsheet_title, spreadsheet_id, google_api_key)
```

```python
@TaskTemplate(iterate_over_data_files=True)
def create_google_single_sheet_urls(sheet_title: str, google_api_key: str) -> HttpUrlModel:
    url = HttpUrlModel('https://sheets.googleapis.com/')
    url.path // 'v4' // 'spreadsheets' // spreadsheet_id // 'values' // sheet_title
    url.query['key'] = google_api_key
    return url
```
